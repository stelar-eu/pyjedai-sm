```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid") 

dataset_type = 'joinable'

main_dir = 'C:/Users/G_A.Papadakis/OneDrive - Public Power Corporation S.A/Schema Matching/Valentine/'
df = pd.read_csv(main_dir + 'results_composite_with_re_pr.csv')
df
```

```python
df[['dataset', 'dataset_type']].value_counts()
```

```python
datasets = df[df['dataset_type'] == dataset_type]
datasets
```

```python
f1 = datasets.filter(regex='F1').copy()
f1
```

```python
f1 = datasets.filter(regex='F1').copy()
f1.drop(columns=['max F1'], inplace=True)
f1.rename(columns={'UNI: F1':'SemU', 'K: F1':'SemK', 'TopKJoin-UNI: F1':'SynU', 'TopKJoin-K: F1':'SynK',
                   'EJoin-UNI: F1':'SimU', 'EJoin-K: F1':'SimK',                
                   'syntactic-UNI: F1':'BlU', 'syntactic-K: F1':'BlK',
                   'CU: F1':'CU', 'CO: F1':'CO','DB: F1':'DB', 'JDM: F1':'JDM', 
                   'SF: F1':'SF', 'EMBDI: F1': 'EmbDI', 'UNICORN: F1':'UNC'}, inplace=True)
cols = list(f1.columns)
i, j = cols.index('SimU'), cols.index('SynK')
cols[i], cols[j] = cols[j], cols[i]
f1 = f1[cols]
f1
```

```python
ax = sns.boxplot(f1)
ax.set_title("F-Measure") 
ax.tick_params(axis='both', labelsize=12)
plt.ylim(-1, 101)
plt.xticks(rotation=45)
plt.savefig(main_dir + 'composite_'+ dataset_type + '_f1')
```

```python
f1.describe()
```
